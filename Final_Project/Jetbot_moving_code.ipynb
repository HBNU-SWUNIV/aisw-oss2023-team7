```python
#jetbot moving only

from jetbot import Robot
import time

robot = Robot()
label_name = #detection

while True:
    if label_name == 'traffic light':
        robot.forward(1)

        if label_name == 'right turn':
            robot.right(0.6)
            time.sleep(0.8)
            continue

        elif label_name == 'left turn':
            robot.left(0.6)
            time.sleep(0.8)
            continue

        elif label_name == 'obstacle':
            robot.left(0.6)
            time.sleep(0.8)
            robot.right(0.6)
            time.sleep(0.8)
            robot.right(0.6)
            time.sleep(0.8)
            robot.left(0.6)
            time.sleep(0.8)
            continue
            
        elif label_name == 'stop sign':
            robot.stop()
            
```

```python
#jetbot moving and buzzer

from jetbot import Robot
import time
from RPi.GPIO as GPIO

robot = Robot()
label_name = #detection

BEEP_pin = 6
GPIO.setmode(GPIO.BCM)
GPIO.setup(BEEP_pin, GPIO.OUT, initial = GPIO.LOW)

GPIO.output(BEEP_pin, GPIO.HIGH)
time.sleep(0.3)
GPIO.output(BEEP_pin, GPIO.LOW)
time.sleep(0.2)

while True:
    if label_name == 'traffic light':
        GPIO.output(BEEP_pin, GPIO.HIGH)
        time.sleep(0.4)
        GPIO.output(BEEP_pin, GPIO.LOW)
        time.sleep(0.2)
        
        robot.forward(1)

        if label_name == 'right turn':
            GPIO.output(BEEP_pin, GPIO.HIGH)
            time.sleep(0.3)
            GPIO.output(BEEP_pin, GPIO.LOW)
            time.sleep(0.2)
            
            robot.right(0.6)
            time.sleep(0.8)
            continue

        elif label_name == 'left turn':
            GPIO.output(BEEP_pin, GPIO.HIGH)
            time.sleep(0.3)
            GPIO.output(BEEP_pin, GPIO.LOW)
            time.sleep(0.2)
            
            robot.left(0.6)
            time.sleep(0.8)
            continue

        elif label_name == 'obstacle':
            GPIO.output(BEEP_pin, GPIO.HIGH)
            time.sleep(0.3)
            GPIO.output(BEEP_pin, GPIO.LOW)
            time.sleep(0.2)
            
            robot.left(0.6)
            time.sleep(0.8)
            robot.right(0.6)
            time.sleep(0.8)
            robot.right(0.6)
            time.sleep(0.8)
            robot.left(0.6)
            time.sleep(0.8)
            continue
            
        elif label_name == 'stop sign':
            GPIO.output(BEEP_pin, GPIO.HIGH)
            time.sleep(0.4)
            GPIO.output(BEEP_pin, GPIO.LOW)
            time.sleep(0.2)
            
            robot.stop()
            
```
